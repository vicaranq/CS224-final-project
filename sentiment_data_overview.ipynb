```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import dataset

SENTIMENT140 = 1
TWITTER = 2
TWITTER_AIRLINES = 3
TWITTER_APPLE = 4
WNUT_2016 = 5
WNUT_2017 = 6
TWEEBANK = 7
```

```python
twitter_train, twitter_validate, twitter_test =  dataset.dataset_reader(TWITTER)
[twitter_train, twitter_validate, twitter_test] = list(map(lambda ds : dataset.prune_columns(2, ds), [twitter_train, twitter_validate, twitter_test]))
```

```python
# Unique values of sentiment
twitter_sentiment_labels = twitter_train['sentiment'].unique()
```

```python
twitter_train_cts = twitter_train['sentiment'].value_counts()
twitter_validate_cts = twitter_validate['sentiment'].value_counts()
twitter_test_cts = twitter_test['sentiment'].value_counts()

X_axis = np.arange(len(twitter_sentiment_labels))
  
plt.bar(X_axis - 0.4, twitter_train_cts, 0.3, label = 'train')
plt.bar(X_axis - 0.1, twitter_validate_cts, 0.3, label = 'validate')
plt.bar(X_axis + 0.2, twitter_test_cts, 0.3, label = 'test')
  
plt.xticks(X_axis, twitter_sentiment_labels)
plt.xlabel("sentiment")
plt.ylabel("frequency")
plt.title("Number of tweets per sentiment")
plt.legend()
plt.show()
```

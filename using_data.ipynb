```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import utils

SENTIMENT140 = 1
TWITTER = 2
TWITTER_AIRLINES = 3
TWITTER_APPLE = 4
WNUT_2016 = 5
WNUT_2017 = 6
TWEEBANK = 7
```

```python
twitter_train, twitter_validate, twitter_test =  utils.dataset_reader(TWITTER)
[twitter_train, twitter_validate, twitter_test] = list(map(lambda ds : utils.prune_columns(2, ds), [twitter_train, twitter_validate, twitter_test]))
```

```python
# Unique values of sentiment
sentiment_labels = twitter_train['sentiment'].unique()
```

```python
train_cts = twitter_train['sentiment'].value_counts()
validate_cts = twitter_validate['sentiment'].value_counts()
test_cts = twitter_test['sentiment'].value_counts()
  
X_axis = np.arange(len(sentiment_labels))
  
plt.bar(X_axis - 0.4, train_cts, 0.3, label = 'train')
plt.bar(X_axis - 0.1, validate_cts, 0.3, label = 'validate')
plt.bar(X_axis + 0.2, test_cts, 0.3, label = 'test')
  
plt.xticks(X_axis, sentiment_labels)
plt.xlabel("sentiment")
plt.ylabel("frequency")
plt.title("Number of tweets per sentiment")
plt.legend()
plt.show()
```
